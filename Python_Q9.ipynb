Write a program to read the data from the following link, perform data analysis and answer the following questions

Link - https://data.wa.gov/api/views/f6w7-q2d2/rows.csv?accessType=DOWNLOAD 

Insights to be drawn -

●         Get all the cars and their types that do not qualify for clean alternative fuel vehicle. 

●         Get all TESLA cars with the model year, and model type made in Bothell City.

●         Get all the cars that have an electric range of more than 100, and were made after 2015.

●         Draw plots to show the distribution between city and electric vehicle type

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the CSV data from the given link
url = "https://data.wa.gov/api/views/f6w7-q2d2/rows.csv?accessType=DOWNLOAD"
data = pd.read_csv(url)
```

```python
data.head()
```

```python
data["Clean Alternative Fuel Vehicle (CAFV) Eligibility"].unique()
```

```python
# Get all the cars and their types that do not qualify for clean alternative fuel vehicle
non_clean_cars = data[data["Clean Alternative Fuel Vehicle (CAFV) Eligibility"] != "Clean Alternative Fuel Vehicle Eligible"]
non_clean_cars_info = non_clean_cars[["Make", "Model"]]
print("Cars and their types that do not qualify for clean alternative fuel vehicle:")
non_clean_cars_info.head()
```

```python
non_clean_cars_info.info()
```

```python
# Get all TESLA cars with the model year, and model type made in Bothell City
tesla_cars = data[(data["Make"] == "TESLA") & (data["City"] == "Bothell")]
tesla_cars_info = tesla_cars[["Model Year", "Electric Vehicle Type"]]
print("\nTESLA cars with the model year, and model type made in Bothell City:")
tesla_cars_info.head()
```

```python
# Get all the cars that have an electric range of more than 100 and were made after 2015
electric_cars = data[(data["Electric Range"] > 100) & (data["Model Year"] > 2015)]
electric_cars_info = electric_cars[["Make", "Model", "Model Year"]]
print("\nCars that have an electric range of more than 100 and were made after 2015:")
electric_cars_info.head()
```

```python
# Draw plots to show the distribution between city and electric vehicle type
plt.figure(figsize=(20, 8))
data["City"].value_counts().plot(kind="bar")
plt.xlabel("City")
plt.ylabel("Number of Electric Vehicles")
plt.title("Distribution of Electric Vehicles by City")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
data["Electric Vehicle Type"].value_counts().plot(kind="bar")
plt.xlabel("Electric Vehicle Type")
plt.ylabel("Number of Electric Vehicles")
plt.title("Distribution of Electric Vehicles by Type")
plt.show()
```
