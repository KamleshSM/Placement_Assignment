# Get all Pokémon whose spawn rate is less than 5%

```python


# Get all Pokémon whose spawn rate is less than 5%
spawn_rate_less_than_5 = df[df['spawn_chance'] < 5]
print("Pokémon with spawn rate less than 5%:")
spawn_rate_less_than_5.head()


```

```python
# Plotting the analysis

# Bar plot for Pokémon spawn rate less than 5%
plt.figure(figsize=(10, 6))
plt.bar(spawn_rate_less_than_5['Name'], spawn_rate_less_than_5['spawn_chance'])
plt.xlabel('Pokémon')
plt.ylabel('Spawn Rate (%)')
plt.title('Pokémon with Spawn Rate Less Than 5%')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
spawn_rate_less_than_5.info()
```

#### Get all Pokémon that do not have more than 2 evolutions

```python
df['Next_evolution'].unique()
```

```python
# Create the "evol1" column by splitting df['Next_evolution'] and counting the total items
df['evol1'] = df['Next_evolution'].str.split(', ').apply(lambda x: len(x) if isinstance(x, list) else 0)

# Replace NaN values in the "evol1" column with 0
df['evol1'].fillna(0, inplace=True)
```

```python
df.head()
```

```python
# Create the "evol1" column by splitting df['Next_evolution'] and counting the total items
df['evol2'] = df['Prev_evolution'].str.split(', ').apply(lambda x: len(x) if isinstance(x, list) else 0)

# Replace NaN values in the "evol1" column with 0
df['evol2'].fillna(0, inplace=True)
```

```python
df.head()
```

```python
df['evol3'] = df['evol2'] + df['evol1']
```

```python
df.head()
```

```python
# Get all Pokémon that do not have more than 2 evolutions
less_than_2_evolutions = df[df['evol3'] >= 2]
less_than_2_evolutions.head()
```

```python
less_than_2_evolutions.info()
```

```python
# Bar plot for Pokémon with less than 2 evolutions
plt.figure(figsize=(10, 6))
plt.bar(less_than_2_evolutions['Name'], less_than_2_evolutions['evol3'])
plt.xlabel('Pokémon')
plt.ylabel('Number of Evolutions')
plt.title('Pokémon with Less Than 2 Evolutions')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

# Insights

1. There are 49 Pokémon that do not have more than 2 evolutions

```python
# Convert spawn time to seconds and get all Pokémon whose spawn time is less than 300 seconds
df['spawn_time'] = pd.to_datetime(df['spawn_time'], format="%M:%S").dt.second
spawn_time_less_than_300 = df[df['spawn_time'] < 300]
print("Pokémon with spawn time less than 300 seconds:")
spawn_time_less_than_300.head()

```

```python
# Bar plot for Pokémon with spawn time less than 300 seconds
plt.figure(figsize=(10, 6))
plt.bar(spawn_time_less_than_300['Name'], spawn_time_less_than_300['spawn_time'])
plt.xlabel('Pokémon')
plt.ylabel('Spawn Time (seconds)')
plt.title('Pokémon with Spawn Time Less Than 300 seconds')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
spawn_time_less_than_300.info()
```

# Insights

1. There are 145 Pokémon with Spawn Time Less Than 300 seconds
